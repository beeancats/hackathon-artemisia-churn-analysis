# Hackathon Artemisia Elas+ Tech  
## Storytelling - Bank Customer Churn

## Contexto do Problema

A evasão de clientes (churn) representa um dos principais desafios estratégicos do setor bancário, impactando diretamente retenção, receita e crescimento das instituições financeiras.

Diante desse cenário, o projeto foi desenvolvido com o objetivo de identificar padrões relacionados ao churn bancário e construir um modelo preditivo capaz de auxiliar na identificação de clientes com maior risco de cancelamento.


## Objetivo do Projeto

Desenvolver uma solução analítica utilizando SQL, Python e Machine Learning para:

- Compreender os fatores associados ao churn;
- Identificar padrões de comportamento dos clientes;
- Prever clientes com maior risco de evasão;
- Apoiar estratégias de retenção orientadas por dados.


## Pipeline ETL

O projeto iniciou com a construção de uma pipeline ETL (Extract, Transform and Load):

### Extract
- Leitura do dataset CSV Bank Customer Churn utilizando Pandas;
### Transform
- Validação de valores nulos;
- Verificação de duplicidades;
- Análise da estrutura dos dados;
- Preparação da base para análise e modelagem;
### Load
- Ingestão dos dados no PostgreSQL para armazenamento estruturado e consultas analíticas.

## Principais Insights
- A variável Exited mostrou que a maior parte dos clientes permaneceu ativa, porém existe uma parcela relevante de evasão;
- Países diferentes apresentaram taxas distintas de churn;
- Clientes menos ativos demonstraram maior tendência ao cancelamento;
- Faixas etárias mais elevadas apresentaram maior risco de churn;
- Variáveis financeiras e comportamentais apresentaram relação com o comportamento de evasão.

## Modelagem Preditiva

Foi utilizado o algoritmo Random Forest para previsão de churn bancário.

O modelo foi escolhido por:

- Apresentar boa performance em problemas de classificação;
- Permitir interpretação da importância das variáveis;
- Possuir boa capacidade de generalização.

## Resultado do Modelo

O modelo apresentou acurácia aproximada de 86%, demonstrando capacidade relevante de identificar clientes com maior risco de churn.

As variáveis mais importantes para o modelo foram:

- idade;
- saldo bancário;
- atividade do cliente;
- quantidade de produtos.

## Recomendações Estratégicas

Com base nas análises realizadas, algumas estratégias podem ser consideradas:

- Criação de campanhas direcionadas para clientes menos ativos;
- Ações específicas para perfis com maior risco de evasão;
- Fortalecimento do relacionamento com clientes de faixas etárias mais críticas;
- Acompanhamento contínuo dos indicadores de churn.

## Storytelling e Visualizações

As visualizações e o storytelling foram desenvolvidos integralmente em Python/Jupyter Notebook, substituindo o uso do Power BI conforme orientação do projeto.

## Conclusão

O projeto demonstrou como a integração entre SQL, Python e Machine Learning pode apoiar decisões estratégicas orientadas por dados.

A solução desenvolvida permitiu transformar dados brutos em insights analíticos e modelos preditivos aplicados ao problema de churn bancário, contribuindo para estratégias de retenção e tomada de decisão baseada em dados.
