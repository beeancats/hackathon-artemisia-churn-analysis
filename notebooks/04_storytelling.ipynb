# Problema de Negócio

A evasão de clientes (churn) representa um dos principais desafios para instituições financeiras, impactando diretamente retenção, receita e crescimento sustentável.

Diante desse cenário, o projeto busca identificar padrões associados ao churn bancário por meio de análise de dados e Machine Learning.

# Arquitetura da Solução

CSV Kaggle
→ PostgreSQL
→ Python
→ SQL
→ EDA
→ Machine Learning
→ Insights Estratégicos

# Principais Insights

- Clientes mais velhos apresentaram maior tendência ao churn;
- Países específicos demonstraram taxas mais elevadas de evasão;
- Clientes menos ativos apresentaram maior risco de cancelamento;
- Variáveis financeiras e comportamentais mostraram forte influência na evasão bancária.

# Resultado do Modelo Preditivo

O modelo Random Forest apresentou capacidade relevante de prever clientes com risco de churn, permitindo identificar padrões importantes para estratégias de retenção.

# Recomendações Estratégicas

Com base nos resultados obtidos, recomenda-se:

- desenvolvimento de campanhas focadas em clientes de maior risco;
- estratégias de retenção para clientes menos ativos;
- acompanhamento de segmentos regionais com maior taxa de churn;
- utilização contínua de modelos preditivos para antecipação de evasão.

# Conclusão

O projeto demonstrou como a integração entre SQL, Python e Machine Learning pode apoiar decisões estratégicas orientadas por dados.

A solução desenvolvida permite identificar padrões de evasão bancária e apoiar ações preventivas de retenção de clientes.
